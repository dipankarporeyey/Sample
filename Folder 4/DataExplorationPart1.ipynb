```python
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Generate some sample data
true_labels = np.array([1, 0, 1, 0, 1, 0, 0, 1, 1, 0])
predicted_scores = np.array([0.8, 0.3, 0.6, 0.2, 0.9, 0.4, 0.1, 0.7, 0.75, 0.25])

# Define a threshold for binary classification
threshold = 0.5

# Apply the threshold to convert predictions into binary classifications
predicted_labels = (predicted_scores >= threshold).astype(int)

# Create a confusion matrix
conf_matrix = confusion_matrix(true_labels, predicted_labels)

# Extract values from the confusion matrix
tn, fp, fn, tp = conf_matrix.ravel()

# Calculate evaluation metrics
accuracy = accuracy_score(true_labels, predicted_labels)
precision = precision_score(true_labels, predicted_labels)
recall = recall_score(true_labels, predicted_labels)
f1 = f1_score(true_labels, predicted_labels)

# Print the confusion matrix and metrics
print("Confusion Matrix:")
print(conf_matrix)
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```
