```python
import os
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
no_of_cluster = 5
```

```python
data = make_blobs(n_samples=300, centers=no_of_cluster, n_features=3)
data = data[0]
data.shape
```

```python
plt.figure(figsize=(14,7))
plt.scatter(data[:,0], data[:,1], color='red')
plt.show()
```

```python
kmeans = KMeans(n_clusters=no_of_cluster)
```

```python
kmeans.fit(data)
```

```python
print(kmeans.cluster_centers_)
print(kmeans.labels_)
```

```python
plt.figure(figsize=(14,7))
for i in range(no_of_cluster):
    plt.scatter(data[kmeans.labels_==i,0], data[kmeans.labels_==i,1], label="Cluster "+str(i+1))
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker="^", s=175, label="Centroid")
plt.legend(loc='upper right')
plt.show()
```
